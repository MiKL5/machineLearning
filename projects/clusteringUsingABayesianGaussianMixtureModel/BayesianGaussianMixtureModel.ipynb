```python
import seaborn           as     sns
from   sklearn.mixture   import BayesianGaussianMixture
import matplotlib.pyplot as     plt
```

# **Partitionner avec un modèle Bayésien Gaussien**

```python
iris = sns.load_dataset('iris')
X    = iris.drop(columns='species')
y    = iris['species']
```

## **Prendre les colonnes '`petal_length`' et '`sepal_length`' et tracer les points de données**

```python
sns.scatterplot(iris, x='petal_length', y='sepal_length', hue='species')
```

## **Créer l’estimateur BayesianGaussianMixture et l'adapter aux Features (X)**

```python
bgmm = BayesianGaussianMixture(n_components=3, random_state=42)
bgmm.fit(X)
y_pred = bgmm.predict(X)
y_pred
```

## **Traçer les clusters**

```python
plt.scatter(X['petal_length'], X['sepal_length'], c=y_pred, cmap='viridis')
```
