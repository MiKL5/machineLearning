```python
import seaborn as sns
from   sklearn.svm import OneClassSVM
```

# **Détecter les anomalies avec OneClassSVM**

```python
df = sns.load_dataset('iris')
df.head()
```

## **Essayer avec '`sepal_width`' et '`petal_width`'**

```python
sns.scatterplot(df, x=df.sepal_width, y=df.petal_width)
```

Il y a quelques outliers.

```python
ocsvm = OneClassSVM(nu=0.001)                  # plus la valeur du paramètre nu est élevée, plus il peut y avaoir de valeurs abérantes
X          = df[['sepal_width', 'petal_width']]
ocsvm.fit(X)
ocsvm.predict(X)
```

```python
sns.scatterplot(df, x=df.sepal_width, y=df.petal_width, hue=ocsvm.predict(X))
```

Trois valeurs abérantes ici si le paramètre '`nu`' est à 0.001. À 0.101, il y en a 11.
