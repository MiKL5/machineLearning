```python
import pandas                  as pd
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot       as plt
from   sklearn.cluster         import AgglomerativeClustering
```

# **Segmentater la clientèle par le clustering hiérarchique**

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
# Utiliser chaque colonnes
features = df.iloc[:, 2:5]
```

```python
# Faire le dendrogramme
dendrogram = sch.dendrogram(sch.linkage(features, method='ward'))
plt.title("Dendrogramme")
plt.xlabel("Clients")
plt.ylabel("Distance Euclidienne")
```

```python
# Mieux voir l'axe X
plt.figure(figsize=(100, 50))
dendrogram = sch.dendrogram(sch.linkage(features, method='ward'))
plt.title("Dendrogramme", fontsize=20)
plt.xlabel("Clients")
plt.ylabel("Distance Euclidienne")
```

Découper en trois clusteurs.

```python
dendrogram = sch.dendrogram(sch.linkage(features, method='ward'))
plt.axhline(y=300, color='r')
plt.title("Dendrogramme")
plt.xlabel("Clients")
plt.ylabel("Distance Euclidienne")
```

```python
cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
cluster.fit_predict(features)
```

```python
plt.scatter(features['Annual Income (k$)'], features['Spending Score (1-100)'], c=cluster.labels_)
plt.xlabel('Revenu annuel')
plt.ylabel('Score de dépenses')
```

Découper en quatre clusters

```python
cluster = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')
cluster.fit_predict(features)
plt.scatter(features['Annual Income (k$)'], features['Spending Score (1-100)'], c=cluster.labels_)
plt.xlabel('Revenu annuel')
plt.ylabel('Score de dépenses')
plt.show()
```

Les revenus élevés sont en vert.
En jaunâtre, les ménages aux revenus modestes, sont dépensiers.
Les clusters du haut (dépensières) sont à viser dans les campagnes d'emailing.
