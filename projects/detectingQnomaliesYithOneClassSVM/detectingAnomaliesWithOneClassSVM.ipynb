```python
import seaborn as sns
from   sklearn.svm import OneClassSVM
```

# **DÃ©tecter les anomalies avec OneClassSVM**

```python
df = sns.load_dataset('iris')
df.head()
```

## **Essayer avec '`sepal_width`' et '`petal_width`'**

```python
sns.scatterplot(df, x=df.sepal_width, y=df.petal_width)
```

Il y a quelques outliers.

```python
ocsvm = OneClassSVM(nu=0.001)                  # plus la valeur du paramÃ¨tre nu est Ã©levÃ©e, plus il peut y avaoir de valeurs abÃ©rantes
X          = df[['sepal_width', 'petal_width']]
ocsvm.fit(X)
ocsvm.predict(X)
```

```python
sns.scatterplot(df, x=df.sepal_width, y=df.petal_width, hue=ocsvm.predict(X))
```
